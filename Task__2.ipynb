```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset

df = sns.load_dataset('titanic')

# Check for missing values

print(df.isnull().sum())

# Handle missing values 

df['age'].fillna(df['age'].median(), inplace=True)

```

```python
# Exploratory Data Analysis
# Visualize distributions of numerical variables

sns.histplot(data=df, x='age', bins=20)
plt.title('Distribution of Age')
plt.show()

sns.histplot(data=df, x='fare', bins=20)
plt.title('Distribution of Fare')
plt.show()
```

```python
# Get summary statistics of numerical variables
print(df.describe())

# Check the data types of variables
print(df.dtypes)

```

```python
# Visualize relationships between categorical variables
sns.countplot(x='survived', data=df)
plt.title('Survived by Gender')
plt.show()

```

```python
sns.countplot(x='survived', hue='class', data=df)
plt.title('Survived by Class')
plt.show()
```

```python
# Visualize relationships between numerical and categorical variables
sns.boxplot(x='survived', y='age', data=df)
plt.title('Survived by Age')
plt.show()
```

```python
sns.boxplot(x='survived', y='fare', data=df)
plt.title('Survived by Fare')
plt.show()
```

```python
# Correlation analysis
correlation = df[['age', 'fare']].corr()
print(correlation)
```

```python
# Cross-tabulation
cross_tab = pd.crosstab(df['class'], df['survived'])
print(cross_tab)
```
